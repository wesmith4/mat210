## Hierarchical clustering 

```python
import numpy as np
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

data = np.array([[5,5,6,8,10]]).T

plt.title("Dendrograms")  
shc.dendrogram(shc.linkage(data, method='single'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
numberOfClusters = 2
cluster = AgglomerativeClustering(n_clusters=numberOfClusters, affinity='euclidean', linkage='ward')  
cluster.fit_predict(data)
for i in range(numberOfClusters):
    index = np.where(cluster.labels_ == i)[0]
    print('Cluster ',i)
    print('  ', data[index].T)
```

### What if I want thicker lines and larger fonts? 

```python
import numpy as np
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

data = np.array([[5,5,6,8,10]]).T

plt.rcParams['lines.linewidth'] = 3

fig = plt.figure(figsize=(10, 7))  
ax = fig.add_subplot(1,1,1)
plt.title("Dendrograms",fontsize=20)  
dend = shc.dendrogram(shc.linkage(data, method='single'))
ax.tick_params(axis='x', which='major', labelsize=15)
ax.tick_params(axis='y', which='major', labelsize=15)
plt.show()
```

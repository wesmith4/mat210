```python
import numpy as np
import pandas as pd
```

```python
# Create G matrix from the network
G = np.array([[.15/6,.15/6,.15/6,.15/6+.85,.15/6,.15/6+.85],
              [.15/6+.85,.15/6,.15/6,.15/6,.15/6,.15/6],
              [.025,.450,.025,.025,.025,.025],
              [.025,.450,.85/3+.15/6,.025,.025,.025],
              [.025,.025,.85/3+.15/6,.025,.025,.025],
              [.025,.025,.85/3+.15/6,.025,.875,.025]])
```

```python
# Create initial vector
v0 = np.array([1,0,0,0,0,0]).reshape((6,1))
```

```python
PageRatings = np.zeros((20,6))
Rankings = np.zeros((20,6))
PageRatings[0] = v0.reshape((1,6))

for n in range(1,20):
  PageRatings[n] = np.matmul(G,PageRatings[n-1].reshape((6,1))).reshape(1,6)
  Rankings[n] = np.argsort(-PageRankings[n],axis=0)
print(PageRatings)
print(Rankings)
```

```python
evals,evecs = np.linalg.eig(G)
```
